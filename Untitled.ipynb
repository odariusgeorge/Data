# DATA HANDLER

```python
left_join = pd.merge(left=genesData, right=countriesData, how='left', left_on='Region/Haplogroup', right_on='Country')
```

```python
left_join.head()
```

```python
left_join = left_join.fillna(0) #no deaths
```

```python
left_join = left_join.drop(left_join[left_join.Total < 250].index)
```

```python
left_join.head()
```

```python
left_join = left_join.drop(['Country',
                'Tot_Cases_\n1M_pop','Total'], axis=1) #country,total_cases,total ignored for the moment
```

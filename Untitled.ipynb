```python
from sklearn.feature_selection import RFE

rf_small = RandomForestRegressor(n_estimators=100)
selector = RFE(rf_small, 
                 step=1, # features removed at each step
                 n_features_to_select=1 # selected features
                )
selector.fit(X,y)
```

```python
def feature_el(i):
    selector = RFE(rf, i, step = 1)
    selector = selector.fit(X,y)
    selector.support_
    is_approved = lambda i : selector.support_[i] == True
    selected_features = [x for x in features if is_approved(features.index(x))]
    return (selected_features)
```
